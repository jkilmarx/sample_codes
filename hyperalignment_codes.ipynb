# Hyperfing Codes
***

The following script shows how to use the functions provided in __hyperalignment_functions.py__

First, import the hyperalignment_functions script along with the other libraries needed for this analysis.


```python
import hyperalignment_functions as hf

import os
import numpy as np
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt


%matplotlib inline
```

Next, save the full path for the directory with subject data. Use __os.listdir__ to save the subject names into a list. Make sure to remove the anatomical mask file to the list. Finally, use the __get_data__ function in hyperalignment_functions.py to load all the fMRI and behavioral data into a pandas dataframe. 

```python
subdir = '/media/justin/Seagate Backup Plus Drive/Justin_data/healthy_data'
sub_names = os.listdir(subdir)
sub_names = sub_names[:-1]

data = hf.get_data(sub_names=sub_names, subdir=subdir, timer=True)
```

## Within-subject analysis

The __within_score__ function can be used to perform a leave-one-run-out cross_validation within a single subject's native space. This function has an optional *session* argument that can be changed to change the numeber of sessions depending on which subject you are testing. 

```python
within_scores = hf.within_score(data=data, sub_names=sub_names, plot=True)
```

## Anatomical alignment analysis

First, we need to reload the data using __get_data__, but this time add the argument *alignment=True*.

```python
anatomical_data = hf.get_data(sub_names=sub_names, subdir=subdir, alignment='anatomical', timer=True)
```

The, the __anatomical_score__ function can be used to perform a leave-one-subject-out cross-validation on anatomically aligned data.

```python
anatomical_scores = hf.anatomical_score(data=anatomical_data, sub_names=sub_names, plot=True, timer=True)
```

## Hyperalignment analysis

To perform a leave-one-subject-out cross-validation, use the __cross_validate__ function. This will use 7 runs from all subjects to create the _hyperspace_. These 7 runs will be used for feature selection and hyperalignment. All but 1 subject will be used for classifier training. The left out run for that subject will be used for testing. Scores for each subject are obtained through a leave-one-run-out cross-validation. Predictions are obtained by a linear support vector machine classifier.


```python
hyper_scores = hf.cross_validate(data=data, sub_names=sub_names, plot=True)
```

```python
print('anatomical: ', round(np.mean(anatomical_scores)*100, 2))
print('hyperalignment: ', round(np.mean(hyper_scores)*100, 2))
print('within: ', round(np.mean(within_scores)*100, 2))

group = [['anatomical']*len(sub_names), ['hyperalignment']*len(sub_names), ['within']*len(sub_names)] 
group = np.hstack(group)
scores = [anatomical_scores, hyper_scores, within_scores]
scores = np.hstack(scores) * 100

df = {'group': group, 'score': scores}
df = pd.DataFrame(df)

colors = [sea.color_palette()[1], sea.color_palette()[0], sea.color_palette()[2]]
      
      
plt.figure(figsize=(5,2.5), dpi=500)
sea.set_context("paper", font_scale=1.5)
sea.barplot(x='group', y='score', palette=colors, data = df)
plt.ylim(0,124)
plt.xlabel('')
plt.ylabel('Classification Accuracy')
```

Each of these steps in __cross_validate__ can also be done separately. Use __select_features__ to perform feature selection on the desired runs. These features can then applied to all runs using __apply_features__. This will add a new column of _features_ to the dataframe.

```python
total_runs = np.arange(8)
train_runs = np.arange(7)
test_run = [7]

feat_trans = hf.select_features(data=data, sub_names=sub_names, runs1=train_runs, runs2=train_runs)
feat_train = hf.apply_features(data=data, transform=feat_trans, sub_names=sub_names, runs1=total_runs, runs2=total_runs)

feat_train.head()
```

The __create_common_model__ function can be used to create the *hyperspace*. This function will hyperalign subjects togeter according to Haxby et al. 2011. This function is set to return the transformation matrices for each subject into *hyperspace* by default, but can also be set to return the *hyperspace* template. The __hyperalign__ function can be used to apply the transformation matrices to all runs. This will add a new column named *fmri_hyper* to the dataframe.

```python
hyper_trans = hf.create_common_model(data=feat_train, sub_names=sub_names, runs1=train_runs, runs2=train_runs)
hyper_train = hf.hyperalign(data=feat_train, transform=hyper_trans, sub_names = sub_names, runs1=total_runs, runs2=total_runs)

hyper_train.head()
```

The __hyper_score__ function can be called to obtain classification scores for each subject. This function will train and test a SVM classifier on the specified runs for each subject.

```python
score = hf.hyper_score(data=hyper_train, sub_names=sub_names, train_runs1=train_runs, train_runs2=train_runs, test_runs1=test_run, test_runs2=test_run)
print(score)
print(round(np.mean(score), 2)*100)
```

## Dropping fingers
***
You can also use the optional *drop_fingers* argument in these functions to remove that finger from the dataset. The easiest way is to drop in the __cross_validate__ function. The index, middle, ring, and little fingers can be referenced by [0], [1], [2], and [3], respectively. Multiple fingers can also be dropped, such as [0,1]. The default value to not drop any fingers is [5]. The optional *get_matrix* argument will return a matrix of prediction scores for each finger (like the values in the confusion matrices). This is useful for plotting.

```python
drop_fing = [[5], [0], [1], [2], [3]]
all_matrix = []
scores = []
for drop in drop_fing:
    matrix = hf.cross_validate(data=data, sub_names=sub_names, drop_fingers=drop, get_matrix=True)
    score = hf.cross_validate(data=data, sub_names=sub_names, drop_fingers=drop, plot=True)
    all_matrix.append(matrix)
    scores.append(score)
```

```python
index_in = all_matrix[0][0::4]
print('index in: ', round(np.mean(index_in)*100, 2))
index_out = all_matrix[1][0::4]
print('index out: ', round(np.mean(index_out)*100,2))
middle_in = all_matrix[0][1::4]
print('middle in: ', round(np.mean(middle_in)*100, 2))
middle_out = all_matrix[2][1::4]
print('middle out: ', round(np.mean(middle_out)*100, 2))
ring_in = all_matrix[0][2::4]
print('ring in: ', round(np.mean(ring_in)*100, 2))
ring_out = all_matrix[3][2::4]
print('ring out: ', round(np.mean(ring_out)*100, 2))
little_in = all_matrix[0][3::4]
print('little in: ', round(np.mean(little_in)*100, 2))
little_out = all_matrix[4][3::4]
print('little out: ', round(np.mean(little_out)*100 ,2))

all_fings = [index_in, index_out, middle_in, middle_out, ring_in, ring_out, little_in, little_out]
all_fings = np.hstack(all_fings) *100

finger = [['index']*2*len(sub_names), ['middle']*2*len(sub_names), ['ring']*2*len(sub_names), ['little']*2*len(sub_names)]
finger = np.hstack(finger)
group = [['included']*len(sub_names), ['dropped']*len(sub_names)] * 4
group = np.hstack(group)

df = {'finger': finger, 'group': group, 'score': all_fings}
df = pd.DataFrame(df)

colors = [sea.color_palette()[0], sea.color_palette()[3]]

plt.figure(figsize=(5,2.5), dpi=500)
sea.set_context("paper", font_scale=1.2)
sea.barplot(x='finger', y='score', hue='group', data = df, palette = colors, ci='sd')
plt.ylim(0,124)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('')
plt.ylabel('Classification Accuracy')
```

## Real-time simulation
***
There are 2 sessions of data for subjects ff001-ff006, so it is possible to simulate a real-time environment. The following code block creates the *hyperspace* using the first session data in all but the test subject. The classifier is then trained on this data. Next, the first session of the test subject is aligned to the *hyperspace* and tested on the second session to see a baseline, full session result. Then, scoring of this second session is tested in a simulated real-time environment. The data is loaded one run at a time and goes through a couple pre-processing steps. This data can be z-scored using just the baseline data (first 20 TRs) or real-time z-scoring (up to current TR). The feature selection is then applied to this real-time data and it is aligned to the *hyperspace* and scored. 

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from nilearn.input_data import NiftiMasker

train_session = 1
train_runs = np.arange(8)

scores1=[]
means = []
for test_sub in sub_names[0:6]:

    train_subs = []
    for i in sub_names:
        if (test_sub != i):
            train_subs.append(i)
    
    # create the hyperspace
    train_data = data[data.subject != test_sub]
    feat_trans = hf.select_features(data=train_data, sub_names=train_subs, runs1=train_runs, runs2=np.arange(8))
    feat_train = hf.apply_features(data=train_data, transform=feat_trans, sub_names=train_subs, runs1=train_runs, runs2=np.arange(8))
    hyper_trans = hf.create_common_model(data=feat_train, sub_names=train_subs, runs1=train_runs, runs2=np.arange(8))
    common_model = hf.create_common_model(data=feat_train, sub_names=train_subs, runs1=train_runs, runs2=np.arange(8), get_template=True)
    hyper_train = hf.hyperalign(data=feat_train, transform=hyper_trans, sub_names=train_subs, runs1=train_runs, runs2=np.arange(8))

    # train the classifier
    fmri_train = hyper_train['fmri_hyper'].values.tolist()
    conditions_train = hyper_train['probe'].values
    svc = SVC(kernel='linear')
    svc.fit(fmri_train, conditions_train)

    test_data = data[data.subject == test_sub]
    feat_trans2 = hf.select_features(data=test_data, sub_names=[test_sub], runs1=np.arange(8), runs2=np.arange(8))
    feat_test = hf.apply_features(data=test_data, transform=feat_trans2, sub_names=[test_sub], runs1=np.arange(16), runs2=np.arange(8))

    # align test subject to hyperspace and get baseline score
    hyper_trans2 = hf.align_with_model(data=feat_test, template=common_model, sub_names=[test_sub], runs1=np.arange(8), runs2=np.arange(8))
    hyper_test = hf.hyperalign(data=feat_test, transform=hyper_trans2, sub_names=[test_sub], runs1=np.arange(16), runs2=np.arange(8))
    hyper_test = hyper_test[hyper_test.session == '2']
    fmri_test = hyper_test['fmri_hyper'].values.tolist()
    conditions_test = hyper_test['probe'].values
    pred = svc.predict(fmri_test)
    score = accuracy_score(conditions_test, pred)
    scores1.append(score)
    print('%s baseline session score: %0.5s' %(test_sub, score*100))

    # compute real-time hyperalignment score
    if train_session == 1:
        behavioral = pd.read_csv(("/media/justin/Seagate Backup Plus Drive/Justin_data/fingfind/%s/ref/ft-data-sess2.txt" %test_sub))
        behavioral['session'] = '2'
        behavioral['subject'] = test_sub
        behavioral = behavioral[::10]
    if train_session == 2:
        behavioral = pd.read_csv(("/media/justin/Seagate Backup Plus Drive/Justin_data/fingfind/%s/ref/ft-data-sess1.txt" %test_sub))
        behavioral['session'] = '1'
        behavioral['subject'] = test_sub
        behavioral = behavioral[::10]

    scores2 = []
    for j in range(1,9):

        mask_file = '/media/justin/Seagate Backup Plus Drive/Justin_data/fingfind/%s/ref/mask_lh_m1s1.nii' %test_sub
        if train_session == 1:
            run_file = '/media/justin/Seagate Backup Plus Drive/Justin_data/fingfind/%s/bold/sess2/rrun-00%s.nii' %(test_sub,j)
        if train_session == 2:
            run_file = '/media/justin/Seagate Backup Plus Drive/Justin_data/fingfind/%s/bold/sess1/rrun-00%s.nii' %(test_sub,j)

        masker = NiftiMasker(mask_img=mask_file, detrend=True, t_r=2)
        ds = masker.fit_transform(run_file)

        clf_out = []
        for i in range(8,len(ds[20:])+1,8):
            fmri_masked = ds[:(20+i)]

            # eliminate TRs 1,2,3,7,8 
            a = ([False]*20 + ([False]*3 + [True]*3 + [False]*2)*20)
            three_trs = fmri_masked[a[:(20+i)]]

            # get baseline data
            b = ([True]*20 + ([False]*8)*20)
            baseline = ds[b]

            # average accross TRs 4,5,6
            fmri_data = np.empty((0,len(fmri_masked[1])))
            for i in range(0, len(three_trs), 3):
                temp = np.mean(three_trs[i:i+3],axis=0)
                fmri_data = np.append(fmri_data, [temp], axis=0)

            # real-time zscore
            fmri_zscored = hf.z_score(fmri_masked, fmri_data)

            # baseline zscore
            #fmri_zscored = hf.z_score(baseline, fmri_data)

            fmri_zscored = np.vstack(fmri_zscored)
            temp = pd.DataFrame(fmri_zscored)

            run_labels = behavioral.loc[behavioral['run_num'] == (j-1)]
            data2 = run_labels[:(len(temp))].copy()
            data2['fmri'] = temp.values.tolist()

            features = hf.apply_features(data=data2, transform=feat_trans2, sub_names=[test_sub], runs1=[(j-1)], runs2=0)

            feat_data = features['features'].values.tolist()
            d = np.asmatrix(feat_data)
            fmri_test = (d*hyper_trans2[0]).A
            fmri_test = [fmri_test[-1]]
            conditions_test = features['probe'].values
            conditions_test = [conditions_test[-1]]

            pred = svc.predict(fmri_test)
            score = accuracy_score(conditions_test, pred)
            scores2.append(score)
    mean = np.mean(scores2)
    means.append(mean)
    print('real-time score: %0.5s' %(mean*100))
```

```python
rt_hyper = means
print('Mean real-time hyperalignment score: %0.5s' %(np.mean(rt_hyper)*100))
```
